```python
from grid_generator_latest import generate_grid

json_data = generate_grid(19, b_factor = 5, display_grid = True)
```

```python
row, col = json_data['size']['rows'], json_data['size']['cols']

grid_2d = [json_data['grid'][i:i+col] for i in range(0, len(json_data['grid']), col)]

black_cell_list = []
for i in range(row):
    for j in range(col):
        if grid_2d[i][j] == '.' and grid_2d[row - 1 - i][col - 1 - j]:
            black_cell_list.append([i, j])

print(black_cell_list)
```

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import string
import seaborn as sns

lines = open("./key_log.txt").read().splitlines()

all_keys = set()
keys_pressed = []
for line in lines:
    all_keys.add(line.split("### ")[1])
    keys_pressed.append(line.split("### ")[1])

selected_key = []
for key in all_keys:
    if key.lower() in string.ascii_lowercase:
        selected_key.append(key.lower())

selected_key = sorted(selected_key)

out_dict = {}
for key in selected_key:
    out_dict[key] = 0

for key in keys_pressed:
    if key.lower() in out_dict.keys():
        out_dict[key.lower()] += 1

labels = list(out_dict.keys())
values = list(out_dict.values())

sns.set_style('whitegrid')
plt.figure(figsize = (10, 5))
plt.bar(labels, values)
plt.xlabel('Letters')
plt.ylabel('Frequency')
plt.title('Frequency of Letters in the Given Data')
plt.show()
```

```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

### Answer-set (unique) outside of our dataset

```python
outside_answer_list = set()
for path in glob.glob("./*.txt"):
    if 'answer' not in path:
        all_answers = open(path).read().splitlines()
        for answer in all_answers:
            if len(answer) >= 2 and len(answer) <= 21:
                outside_answer_list.add(answer)
```

```python
outside_answer_df = pd.DataFrame(outside_answer_list, columns = ['answer'])
outside_answer_df['answer_len'] = outside_answer_df['answer'].apply(len)
```

```python
count_by_len = outside_answer_df.groupby('answer_len').count()
histogram = []
for count in count_by_len.iterrows():
    row = count[0]
    if count[0] >= 2 and count[0] <= 21:
        histogram.append([count[0], count[1]['answer']])

histogram = np.array(histogram)

x_values = histogram[:, 0]
y_values = histogram[:, 1]

sns.set_style('whitegrid')
plt.figure(figsize = (10, 5), dpi = 90)
plt.bar(x_values, y_values, width = 0.75)
plt.xticks(x_values)
plt.xlabel('Answer - Lengths')
plt.ylabel('Counts')
plt.show()
```

### Gathering new answer-set

```python
all_words = set()
for path in glob.glob('./*.txt'):
    words = open(path).read().splitlines()
    for word in words:
        all_words.add(word)
```

```python
all_words = list(all_words)
answer_df = pd.DataFrame(all_words, columns = ['answer'])
answer_df['answer_len'] = answer_df['answer'].apply(len)
```
